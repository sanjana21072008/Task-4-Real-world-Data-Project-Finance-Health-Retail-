```python
import pandas as pd

data = {
    'Product':['Pen','Book','Pencil','Bag','Notebook','Pen','Book'],
    'Price':[10,100,5,500,50,10,100],
    'Quantity':[2,1,5,1,3,4,2]
}

df = pd.DataFrame(data)

df
```

total sale column

```python
df['Total'] = df['Price'] * df['Quantity']
df
```

best selling product

```python
df.groupby('Product')['Total'].sum()
```

visualization

```python
import matplotlib.pyplot as plt

df.groupby('Product')['Total'].sum().plot(kind='bar')

plt.title("Product Sales Analysis")
plt.show()
```
